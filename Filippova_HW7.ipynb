````python
def prox_alg(f, f_grad, g_prox, x0, num_iter, beta=0.5, fix_lam=None, accel=False):
    conv = [x0]
    x = x0.copy()
    if accel:
        t_prev = 1
        t_next = (1 + np.sqrt(1 + 4 * t_prev**2)) / 2.
    if fix_lam is None:
        lam = 1.
    for i in range(num_iter):
        if accel and i > 0:
            x = x + (t_prev - 1) / t_next * (x - conv[-2])
            t_prev = t_next
            t_next = (1 + np.sqrt(1 + 4 * t_prev**2)) / 2.
        if fix_lam is None:
            while True:
                z = g_prox(x - lam * f_grad(x), lam)
                if f(z) <= f(x) + f_grad(x).dot(z - x) + np.linalg.norm(z - x)**2 / (2 * lam):
                    break
                else:
                    lam = lam * beta
        else:
            z = g_prox(x - fix_lam * f_grad(x), fix_lam)
        x = z.copy()
        conv.append(x)
    return x, conv
````

````python
def huber(x, mu = mu):
    res = np.zeros_like(x)
    res[np.abs(x) <= mu] = x[np.abs(x) <= mu]**2 / (2 * mu)
    res[np.abs(x) >= mu] = np.abs(x[np.abs(x) >= mu]) - mu / 2
    return res

def grad_huber(x, mu = mu):
    res = np.zeros_like(x)
    res[np.abs(x) <= mu] = x[np.abs(x) <= mu] / mu
    res[np.abs(x) >= mu] = np.sign(x[np.abs(x) >= mu])
    return res
````

````python
mus = np.linspace(0.01, 0.5, 4)
def f_1(x):
  return np.linalg.norm(A@x - b)**2 / 2 + gamma*np.sum(huber(x,mus[0]))

def f_2(x):
  return np.linalg.norm(A@x - b)**2 / 2 + gamma*np.sum(huber(x,mus[1]))
def f_3(x):
  return np.linalg.norm(A@x - b)**2 / 2 + gamma*np.sum(huber(x,mus[2]))
def f_4(x):
  return np.linalg.norm(A@x - b)**2 / 2 + gamma*np.sum(huber(x,mus[3]))
````
