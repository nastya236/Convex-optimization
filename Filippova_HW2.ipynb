````python
def SelectStepSize(x, f, h, rho, alpha0, beta1):

    # 0 < rho < 1

    # alpha0 - initial guess of step size

    # beta1 and beta2 - constants from conditions

    alpha = alpha0

    # Check violating sufficient decrease and curvature conditions

    while (f(x - alpha * h) >= f(x) + beta1 * alpha * grad_f(x).dot(h)):

        alpha *= rho

    return alpha
````

````python
def GradientStep(x, f):
    
    h = grad_f(x)
    alpha = SelectStepSize(x , f, h, rho = 0.1, alpha0 = 0.1, beta1 = 0.1)
    x_ = x - alpha * h # так как f дифференцируема по условию, легко посчитаем градиент
    
    return x_
````

````python
def SelectAlpha(A,x):
    alpha = (grad_f(x).T @ grad_f(x))/(grad_f(x).T @ A @ grad_f(x))
    return alpha
````

````python
def SteepestStep(A,b,x):
    alpha = SelectAlpha(A,x)
    
    x_ = x - alpha*grad_f(x)
    
    return x_
````

````python
def binary_search(f, a, b, epsilon, callback=None):
    c = (a + b) / 2.0
    while abs(b - a) > epsilon:
#         Check left subsegment
        y = (a + c) / 2.0
        if f(y) <= f(c):
            b = c
            c = y
        else:
#         Check right subsegment
            z = (b + c) / 2.0
            if f(c) <= f(z):
                a = y
                b = z
            else:
                a = c
                c = z
        if callback is not None:
            callback(a, b)
    return c
````

````python
def golden_search(f, a, b, tol=1e-5, callback=None):
    tau = (np.sqrt(5) + 1) / 2.0
    y = a + (b - a) / tau**2
    z = a + (b - a) / tau
    while b - a > tol:
        if f(y) <= f(z):
            b = z
            z = y
            y = a + (b - a) / tau**2
        else:
            a = y
            y = z
            z = a + (b - a) / tau
        if callback is not None:
            callback(a, b)
    return (a + b) / 2.0
````
